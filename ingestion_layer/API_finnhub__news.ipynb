```python
import requests
import pandas as pd
import time
from datetime import date, datetime

# Carica lista dei simboli S&P 500
url = "https://raw.githubusercontent.com/datasets/s-and-p-500-companies/master/data/constituents.csv"
tickers = pd.read_csv(url)['Symbol'].tolist()

# Parametri comuni
API_KEY = "d056jb9r01qoigrsmf5gd056jb9r01qoigrsmf60"
today = date.today().strftime("%Y-%m-%d")
FROM_DATE = TO_DATE = today

print(FROM_DATE)
# Lista per salvare tutte le notizie
news_data = []

# Loop su tutti i simboli
for i, symbol in enumerate(tickers):
    url = f"https://finnhub.io/api/v1/company-news?symbol={symbol}&from={FROM_DATE}&to={TO_DATE}&token={API_KEY}"
    
    try:
        response = requests.get(url)
        data = response.json()

        print(f"[{i+1:03d}] {symbol}: {len(data)} notizie trovate")

        
        for news in data:
            news_data.append({
                "date": pd.to_datetime(news['datetime'], unit='s'),
                "symbol_requested": symbol,
                "related": news.get("related", ""),
                "headline": news.get("headline", ""),
                "summary": news.get("summary", ""),
                "source": news.get("source", ""),
                "url": news.get("url", ""),
                "image": news.get("image", ""),
                "id": news.get("id", None)
            })


    except Exception as e:
        print(f"Errore su {symbol}: {e}")

    time.sleep(1)  # evita rate limit

# Converti in DataFrame
df_news_finnhub = pd.DataFrame(news_data)

# Mostra anteprima
print("\n✅ Notizie raccolte:", len(df_news_finnhub))
print(df_news_finnhub.head())


# PROBLEMA: L'API DICE CHE STIAMO MANDANDO TROPPE RICHIESTE, FORSE AUMENTANDO IL TIME SLEEP POSSIAMO
# OVVIARE A QUESTO PROBLEMA. CON UN SECONDO FUNZIONA
```
