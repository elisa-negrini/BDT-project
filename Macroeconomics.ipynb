```python
import requests
import pandas as pd

# GDP: Prodotto Interno Lordo (in dollari correnti). Misura la dimensione dell'economia USA ogni trimestre.
# Trimestrale

# URL API FRED per il GDP
url1 = "https://api.stlouisfed.org/fred/series/observations?series_id=GDP&api_key=119c5415679f73cb0da3b62e9c2a534d&file_type=json"

# Richiesta GET (headers opzionali qui non servono)
response = requests.get(url1)
data = response.json()

# Estrai la lista di osservazioni
observations = data["observations"]

# Crea un DataFrame
df_GDP1 = pd.DataFrame(observations)

# Tieni solo le colonne principali e converti i tipi
df_GDP1 = df_GDP1[["date", "value"]]
df_GDP1["value"] = pd.to_numeric(df_GDP1["value"], errors="coerce")  # "." → NaN

# Ordina per data
df_GDP1["date"] = pd.to_datetime(df_GDP1["date"])
df_GDP1 = df_GDP1.sort_values("date").reset_index(drop=True)

# URL API FRED per il GDP (Percent change from a year ago)
url2 = "https://api.stlouisfed.org/fred/series/observations?series_id=GDP&api_key=119c5415679f73cb0da3b62e9c2a534d&file_type=json&units=pc1"

# Richiesta GET (headers opzionali qui non servono)
response = requests.get(url2)
data = response.json()

# Estrai la lista di osservazioni
observations = data["observations"]

# Crea un DataFrame
df_GDP2 = pd.DataFrame(observations)

# Tieni solo le colonne principali e converti i tipi
df_GDP2 = df_GDP2[["date", "value"]]
df_GDP2["value"] = pd.to_numeric(df_GDP2["value"], errors="coerce")  # "." → NaN

# Ordina per data
df_GDP2["date"] = pd.to_datetime(df_GDP2["date"])
df_GDP2 = df_GDP2.sort_values("date").reset_index(drop=True)

df_GDP1.rename(columns={"value": "value_raw"}, inplace=True)
df_GDP2.rename(columns={"value": "value_pc1"}, inplace=True)

# Merge per data
df_GDP = pd.merge(df_GDP1, df_GDP2, on="date")

# Save Data
df_GDP.to_csv("GDP.csv", index=False)
print("✅ Data saved to 'GDP.csv'")

# Mostra le prime righe
df_GDP.tail()
```
